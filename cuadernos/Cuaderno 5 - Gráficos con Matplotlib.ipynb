### Scatter Plot ( o nube de puntos) 

Para hacer un grfico de tipo nube de puntos, vamos a utilizar el mtodo `.scatter()` de matplotlib.

- x, y: Un par de listas de valores para las coordenadas x e y de cada uno de los puntos a representar
- c: Para cambiar el color del relleno
- edgecolors: Para cambiar el color del contorno
- alpha: Para cambiar la transparencia del punto 
- marker: Para cambiar la forma de punto o, x, >, <
- s: Para cambiar el tamao de los puntos
- linewidths: Para cambiar el grosor del contorno




```python
import matplotlib.pyplot as plt
import numpy as np

#representar las alturas y pesos de cada uno de los alumnos de la clase
alturas=np.array([1.8, 1.78, 1.8, 1.84, 1.6, 2.1, 1.9 ])
pesos=np.array([71, 74, 83, 79, 56, 109, 103 ])

plt.title("Relacin altura/peso de los alumnos de PIA")
plt.xlabel("Alturas de los alumnos")
plt.ylabel("Peso de los alumnos")
plt.scatter(alturas,pesos,s=200,c='red',edgecolors='blue',alpha=1,marker='X',linewidths=3)
plt.show()
```

**Observacin**: Puede utilizar scatter tantas veces como quiera en un grfico

```python
# grfico que representa la funcin matemtica f(x)=2x+1 y la funcin x+1
x=np.arange(-100,100,1)
y=2*x+1
z=x**2+1

plt.title("f(x)=2x+1 vs f(x)=x+1")
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(x,y,marker='o',c='seagreen')
plt.scatter(x,z,marker='x',c='orange')
plt.show()
```

```python
# Puedo jugar con diferentes tamaos y colores

x= [1,2,3,4,5,6,7]
colores=['red','blue','orange','yellow','green','cyan','magenta']
tams=[100,150,200,250,300,350,400]

plt.title("Jugando con diferentes tamaos y colores")
plt.scatter(x,x,c=colores,s=tams)
plt.show()
```

## Line plots (o plot)
Son grficos para pintar lneas contnuas, con el mtodo `plot()`.

Algunos parmetros de este mtodo:
- x, y : Coordenadas del eje x y del eje y de los puntos que forman la lnea. 
- color: Cambiar el color del relleno
- fmt (format): Para establecer el formato bsico como si fuera un string. Por ejemplo: "or" son crculos rojos. El formato es `[marker][linestyle][color]` aunque tambin permite `[color][marker][linestyle]`
- linestyle (ls): Cambia el estilo de la lnea 
- linewidth (ln): Cambia el ancho de la lnea
- markersize: Cambiar el tamao de los puntos de la lnea
- markeredgecolor: Cambiar el color del borde del marcador
- markerfacecolor: Cambia el color del relleno del punto

Ejemplos de fmt:
- 'b'    # blue markers with default shape
- 'or'   # red circles
- '-g'   # green solid line
- '--'   # dashed line with default color
- '^k:'  # black triangle_up markers connected by a dotted line




```python
x=[1,10]
y=[-10,10]


plt.title("Ejemplo de line plot")
plt.plot(x,y,'bo-')
plt.show()
```

```python

#representar las alturas y pesos de cada uno de los alumnos de la clase
alturas=np.array([1.8, 1.78, 1.8, 1.84, 1.6, 2.1, 1.9 ])
pesos=np.array([71, 74, 83, 79, 56, 109, 103 ])

plt.title("Relacin altura/peso de los alumnos de PIA")
plt.xlabel("Altura en metros")
plt.ylabel("Peso en kg")
plt.plot(sorted(alturas),pesos,"*--c",lw=2,markersize=20,mfc="m")
plt.show()

```

## Grfico de barras o Bar Plot

Para hacer un grfico de barras, utilizar el mtodo .bar() que nos da un grfico de barras verticales.
Si queremos barras horizontales usamos el mtodo .barh()

Algunos parmetros de estos grficos:
- x: lista de floats que indica las categoras o coordenadas de las barras
- height: Cambiar la altura de cada una de las barras
- width: Para cambiar la anchura de las barras
- linewidth: Para cambiar el ancho de las lneas
- bottom: Para cambiar el valor vertical mnimo de las barras
- align: Posicin de la marca del eje horizontal con respecto a la base de las barras. Por defecto es 'center'
- otros ya visto como edgecolor, color, etc.




```python
# Sacar un grfico de barras con el nmero de suspensos, aprobados, notables y sobresalientes

notas=['suspenso','aprobado','notable','sobresaliente']
cuenta=[1,10,6,10]

#modificar el alto y ancho de una figura
plt.figure(figsize=(5,3))

plt.bar(notas,cuenta,color=["#E19696","#96E1AF","#516CF0","#002BFF"],width=0.2)
plt.xlabel("Notas")
plt.ylabel("Nmero de alumnos")
plt.show()
```

```python
# Sacar un grfico de barras horizontales con el nmero de suspensos, aprobados, notables y sobresalientes

notas=['suspenso','aprobado','notable','sobresaliente']
cuenta=[1,10,6,10]

#modificar el alto y ancho de una figura
plt.figure(figsize=(5,3))

plt.barh(notas,cuenta,color=["#E19696","#96E1AF","#516CF0","#002BFF"])
plt.xlabel("Notas")
plt.ylabel("Nmero de alumnos")
plt.show()
```

## Grficos circulares (o tarta), Pie Charts

Hay que utilizar el mtodo `pie()`
parmetros interesantes:
- x: Una lista de valores
- labels: Una lista con las etiquetas de cada sector
- colors: Una lista de colores para cada sector
- autopct: Para editar el formato en que se muestra el porcentaje de rea
- labeldistance: Para ajustar la distancia radial de las etiquetas
- radius: Modificar el radio del grfico
- startangle: Cambiar el ngulo (en grados) con el que se empieza
- explode: Lista de valores para indicar la distancia al centro de cada uno de los sectores
- shadow: Aadir sombra



```python
# Representa un grfico de tipo tarta para saber a cuntos les gusta la pizza con pia o sin pia

opciones=['Con pia','sin pia']
cuenta=[20,80]
plt.pie(cuenta,colors=['yellow','cyan'],labels=opciones)
plt.title("La pizza, con pia o sin pia?")
plt.show()
```

```python
# Representa un grfico de tipo tarta para saber a cuntos les gusta la pizza con pia o sin pia
# con parmetros autopct, startangle, explode, shadow
opciones=['Con pia','sin pia']
cuenta=[21.29,100-21.29]
plt.pie(cuenta,colors=['yellow','cyan'],labels=opciones,autopct="%0.2f%%",
        startangle=90,explode=[0.2,0],shadow=True,
       wedgeprops = {'linewidth': 3})
plt.title("La pizza, con pia o sin pia?")
plt.show()
```

## Grficos mltiples

Hacer varios grficos dentro de uno. Vamos a utilizar dos opciones:

1. Hacer uso de la funcin subplots y luego subplot.
  - Creamos un conjunto de grficos ordenados en filas y columnas (subplots) y seleccionamos grfico a grfico con el mtodo subplot. <-> Creamos los contenidos que queramos en cada uno de los grficos
  
  - El mtodo subplots, devuelve una tupla con la figura, y otra tupla con cada uno de los grficos. Vamos a modificar cada uno de los objetos de la segundo tupla para darle el contenido que queramos
  

```python
# 1 forma: Crear un grfico en 1 fila y dos columnas
fig, ejes=plt.subplots(1,2, figsize=(10,7))

plt.subplot(1,2,1)  #(numero filas, numero columnas, indice) #selecciono el grfico 1 de la grid (1,2)

puntos_x=[-10,10]
puntos_y=[20,30]
plt.plot(puntos_x,puntos_y,c='b',linewidth=2)

plt.subplot(1,2,2)
plt.scatter(puntos_x,puntos_y,s=[100,200])

plt.show()

```

```python
# Crear un grfico con 4 subgrficos: 
#    En el primero, dibujad la lnea con relacin altura/peso
#    En el segundo grfico, dibujad la funcin x+1
#    En el tercero, grfico circular pia vs no pia
#    En el cuarto, histograma con la distribucin de pareto

fig,ejes=plt.subplots(2,2,figsize=(15,10))


plt.subplot(2,2,1)
#para el grfico 1: #representar las alturas y pesos de cada uno de los alumnos de la clase
alturas=np.array([1.8, 1.78, 1.8, 1.84, 1.6, 2.1, 1.9 ])
pesos=np.array([71, 74, 83, 79, 56, 109, 103 ])
plt.plot(sorted(alturas),pesos,"*--b",lw=2,markersize=20,mfc='m')
plt.title("Relacin altura/peso")

#para el grfico 2: 
plt.subplot(2,2,2)
# Representa un grfico que muestre las funciones matemticas f(x)=2x+1 y la funcin x+1 con "plot"
x=np.arange(-100,100,20)
y=2*x+1
z=x**2+1
plt.scatter(x,y,marker='o',c='green')
plt.scatter(x,z,marker='x',c='orange')
plt.title("funciones")


#para el grfico 3: 
plt.subplot(2,2,3)
opciones=['Con pia','sin pia']
cuenta=[21.29,100-21.29]
plt.pie(cuenta,labels=opciones,shadow=True,explode=[0.2,0])
plt.title("Pia vs No pia")


#para el grfico 4: 
plt.subplot(2,2,4)
data=np.random.pareto(7.0,2000)
bins=50
plt.hist(data,bins=bins)
plt.title("otro")

plt.suptitle("Grfico de grficos")

plt.show()

```

```python
#2 forma, utilizando los objetos ax
fig,(eje1,eje2)=plt.subplots(2,2,figsize=(15,10))

ax1,ax2=eje1
ax3,ax4=eje2

#para el grfico 1: #representar las alturas y pesos de cada uno de los alumnos de la clase
alturas=np.array([1.8, 1.78, 1.8, 1.84, 1.6, 2.1, 1.9 ])
pesos=np.array([71, 74, 83, 79, 56, 109, 103 ])
ax1.plot(sorted(alturas),pesos,"*--b",lw=2,markersize=20,mfc='m')
ax1.set_title("Relacin altura/peso")

#grfico 2
x=np.arange(-100,100,20)
y=2*x+1
z=x**2+1
ax2.scatter(x,y,marker='o',c='green')
ax2.scatter(x,z,marker='x',c='orange')
ax2.set_title("funciones")

#grfico 3
opciones=['Con pia','sin pia']
cuenta=[21.29,100-21.29]
ax3.pie(cuenta,labels=opciones,shadow=True,explode=[0.2,0])
ax3.set_title("Pia vs No pia")

#grfico 4
data=np.random.pareto(7.0,2000)
bins=50
ax4.hist(data,bins=bins)
ax4.set_title("otro")


plt.show()
```

### Cmo generar grficos de distinto tamao?



Hay que recurrir al mtodo `add_gridspec()` que recibe cuntas filas y cuantas columnas van a disponer el layout del grfico. 

Con el mtodo .add_subplot vamos a indicar de esa especificacin anterior, cuntas filas y cuantas columnas va a ocupar el grfico.


```python
# Creamos el objeto figure
fig = plt.figure(constrained_layout = True)

# Configuramos que tendr 3 filas y 4 columnas
gs = fig.add_gridspec(3, 4)

# Creamos los subplots. En este caso 3
# * Ocupa todas las filas y todas las columnas salvo la ltima
fig_ax1 = fig.add_subplot(gs[:, :-1])
fig_ax1.set_title("gs[:, :-1]")
#para el grfico 1: #representar las alturas y pesos de cada uno de los alumnos de la clase
alturas=np.array([1.8, 1.78, 1.8, 1.84, 1.6, 2.1, 1.9 ])
pesos=np.array([71, 74, 83, 79, 56, 109, 103 ])
fig_ax1.plot(sorted(alturas),pesos,"*--b",lw=2,markersize=20,mfc='m')


# * Ocupa todas las filas salvo la ltima y la ltima columna
fig_ax2 = fig.add_subplot(gs[:-1, -1])
fig_ax2.set_title("gs[:-1, -1]")
x=np.arange(-100,100,20)
y=2*x+1
z=x**2+1
fig_ax2.scatter(x,y,marker='o',c='green')
fig_ax2.scatter(x,z,marker='x',c='orange')
fig_ax2.set_title("funciones")

# * Ocupa la ltima fila y la ltima columna
fig_ax3 = fig.add_subplot(gs[-1, -1])
fig_ax3.set_title("gs[-1, -1]")
opciones=['Con pia','sin pia']
cuenta=[21.29,100-21.29]
fig_ax3.pie(cuenta,labels=opciones,shadow=True,explode=[0.2,0])
fig_ax3.set_title("Pia vs No pia")


plt.show()
```
